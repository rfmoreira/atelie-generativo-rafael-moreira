Usem o modelo BLIP para gerar rascunhos automáticos e revisem manualmente —
a revisão humana é obrigatória e será verificada.

```python
!pip -q install transformers pillow
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image

proc = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
blip = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base").to("cuda")

img = Image.open("dados/img_001.jpg").convert("RGB")

saida = blip.generate(**proc(img, return_tensors="pt").to("cuda"), max_new_tokens=40)

print(proc.decode(saida[0], skip_special_tokens=True)) # Revisar manualmente e prefixar com o token do estilo antes de salvar
```
